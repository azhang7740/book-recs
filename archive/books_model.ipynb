```python
# counts duplicates (no duplicates)
count = 0
for x in range(1, len(short_ratings)):
    if (short_ratings.at[x, 'user_id'] == short_ratings.at[x-1, 'user_id']
        and short_ratings.at[x, 'book_id'] == short_ratings.at[x-1, 'book_id']):
        count = count + 1
count
```

```python
# counts the number of ratings by each user
num_ratings = [0]
user = [1]
index = 0

for x in range(0, len(short_ratings)):
    temp = short_ratings.at[x, 'user_id']
    if temp == user[index]:
        num_ratings[index] = num_ratings[index] + 1
    else:
        num_ratings.append(1)
        user.append(temp)
        index = index + 1
```

```python
count_user_ratings = pd.DataFrame({'user_id': user, 'count_ratings': num_ratings})
count_user_ratings
```
